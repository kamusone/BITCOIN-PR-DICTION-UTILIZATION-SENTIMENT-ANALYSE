# # SCRAPING TWEETS

```python
import tweepy
import re
 
from tweepy import OAuthHandler
#pour se connecter au serveur et r√©cup√©rer les tweet en direct Via tweepy

consumer_api_key = ''
consumer_api_secret = '' 
access_token = ''
access_token_secret =''
```

```python
#indique √† quelle appli se connecter et les token donnent acces
authorizer = OAuthHandler(consumer_api_key, consumer_api_secret)
authorizer.set_access_token(access_token, access_token_secret)
 
```

```python
#R√©cuper les tweets
#si aucun tweet n'est trouv√© au bout de 15 sec , l'appli expire
api = tweepy.API(authorizer ,timeout=15)
 
#creer une liste vide
all_tweets = []
 
search_query = 'bitcoin'
# on utilise l'objet cursor pour pour r√©cuperer les tweets il a plusieurs parametres 
#1 .search: fonction de rechercher 
#3  -filter:retweets : ne recuperer pas les retweet
#4 langue
#5 result_type: que les tweet recents
#6 que les 200 recents tweets
for tweet_object in tweepy.Cursor(api.search,q=search_query+" -filter:retweets",lang='fr',result_type='recent').items(200):
    all_tweets.append(tweet_object.text)
    

 
```

```python
# preprocessing

X = tweets.iloc[:, 10].values  
y = tweets.iloc[:, 1].values
 
 
processed_tweets = []
 
for tweet in range(0, len(X)):  
    # Remove all the special characters
    processed_tweet = re.sub(r'\W', ' ', str(X[tweet]))
 
    # remove all single characters
    processed_tweet = re.sub(r'\s+[a-zA-Z]\s+', ' ', processed_tweet)
 
    # Remove single characters from the start
    processed_tweet = re.sub(r'\^[a-zA-Z]\s+', ' ', processed_tweet) 
 
    # Substituting multiple spaces with single space
    processed_tweet= re.sub(r'\s+', ' ', processed_tweet, flags=re.I)
 
    # Removing prefixed 'b'
    processed_tweet = re.sub(r'^b\s+', '', processed_tweet)
 
    # Converting to Lowercase
    processed_tweet = processed_tweet.lower()
 
    processed_tweets.append(processed_tweet)
 
```

```python
#conversion les texts en chiffre avec TF-IDF

from sklearn.feature_extraction.text import TfidfVectorizer  
tfidfconverter = TfidfVectorizer(max_features=2000, min_df=5, max_df=0.7, stop_words=stopwords.words('french'))  
X = tfidfconverter.fit_transform(processed_tweets).toarray()
```

```python
#former le modele d'analyse sentimentale
from sklearn.ensemble import RandomForestClassifier
text_classifier = RandomForestClassifier(n_estimators=100, random_state=0)  
text_classifier.fit(X, y)
```
